### Mapping for categorical variables

```python
# Mapping for Diabetes
diabetes_mapping = {
    'No': 0,
    'No, pre-diabetes or borderline diabetes': 0,
    'Yes, but female told only during pregnancy': 1,
    'Yes': 1
}
df['Diabetes'] = df['Diabetes'].map(diabetes_mapping)

# One-hot encoding for Sex
df = pd.get_dummies(df, columns=['Sex'])

# Convert remaining categorical variables with "Yes" and "No" values to binary format for correlation computation
binary_columns = ['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Arthritis', 'Smoking_History','Exercise']

for column in binary_columns:
    df[column] = df[column].map({'Yes': 1, 'No': 0})

# Ordinal encoding for General_Health, Checkup, Age_Category
general_health_mapping = {
    'Poor': 0,
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Excellent': 4
}
df['General_Health'] = df['General_Health'].map(general_health_mapping)

checkup_mapping = {
    'Never': 0,
    '5 or more years ago': 0.5,
    'Within the past 5 years': 1,
    'Within the past 2 years': 2,
    'Within the past year': 4

}
df['Checkup'] = df['Checkup'].map(checkup_mapping)

age_category_mapping = {
    '18-24': 0,
    '25-29': 1,
    '30-34': 2,
    '35-39': 3,
    '40-44': 4,
    '45-49': 5,
    '50-54': 6,
    '55-59': 7,
    '60-64': 8,
    '65-69': 9,
    '70-74': 10,
    '75-79': 11,
    '80+': 12
}
df['Age_Category'] = df['Age_Category'].map(age_category_mapping)

# Display the first few rows of the cleaned dataset
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

### FP OR FN?

In the case of heart disease, **false negatives** are often considered more serious than false positives due to the serious consequences that can result from undiagnosed or delayed diagnosis of heart disease. Therefore, healthcare-related machine learning models are often designed to minimize the occurrence of false negatives, even though this may increase the proportion of false positives.

### Analysis report

1. Before Resampling

For the non-heart disease cases (label 0), the model showed high precision (0.92) and an almost perfect recall (1.00), leading to a high F1-score (0.96). This indicates that the model was very effective at identifying individuals without heart disease.
For the actual heart disease cases (label 1), the precision dropped to 0.55 and the recall was alarmingly low (0.06), meaning the model missed 94% of actual heart disease cases, resulting in a very low F1-score (0.11).
This suggests that before resampling, the model had a significant weakness in detecting the minority class (heart disease patients), which is a common issue in imbalanced datasets: a tendency to predict the majority class.

2. After Resampling

For label 0 predictions, precision slightly decreased to 0.98, but the recall significantly dropped to 0.73, with a corresponding decrease in the F1-score to 0.84. While precision remains high, the drop in recall indicates the model began misclassifying more non-heart disease cases as having the disease.
For label 1 predictions, precision decreased to 0.21, but recall substantially increased to 0.79, with an improved F1-score of 0.33. This shows that the model, after addressing the class imbalance, was able to identify most of the actual heart disease cases, albeit at the cost of misclassifying more healthy cases as heart disease.
After addressing the class imbalance, there's a decrease in overall accuracy (from 0.92 to 0.74), due to the model making more errors in identifying non-heart disease cases. However, in heart disease prediction, as discussed earlier, reducing false negatives (increasing recall) might be more critical than maintaining high accuracy.

**Conclusion**

In summary, after addressing class imbalance, the model performs better at identifying actual heart disease cases, despite generating more false positives. In practical applications, there may need to be further model adjustments or the use of cost-sensitive learning approaches to balance precision and recall for optimal diagnostic outcomes.

Moreover, based on the recall ad accuracy for each model after resample the training dataset, we can find ensemble model achieves a googd balance between recall and accuracy. And models with higher recall tend to have lower accuracy.
